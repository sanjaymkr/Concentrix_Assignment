```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
colors = ['red', 'green', 'blue', 'purple']
plt.figure(figsize=(10,5))
bar_width = 0.2
plt.bar(df['total_experience'], df['candidate_name'], color=colors, width=bar_width)
tick_positions = np.arange(0, len(df['total_experience'])*1.2)
plt.xticks(tick_positions, tick_positions)
plt.title('Candidate Experience')
plt.xlabel('Total Experience (Years)')
plt.ylabel('Candiadate Name')
plt.show()
```
